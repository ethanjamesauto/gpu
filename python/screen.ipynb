```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = open('in.txt')
screen = np.empty((256*256, 3))
i = 0
for line in file.readlines():
    nums = line.split('\t')
    for j in range(1, 17):
        #data - rrgggbba
        screen[i][0] = ((int(nums[j], 16) >> 6) &  0b11) / 3
        screen[i][1] = ((int(nums[j], 16) >> 3) & 0b111) / 7
        screen[i][2] = ((int(nums[j], 16) >> 1) & 0b11) / 3
        #screen[i][0] = int(nums[j], 16) / 256
        i += 1
screen = screen.reshape((256, 256, 3))
```

```python

plt.figure(figsize=(2,2), dpi=320)
plt.axis('off')
plt.imshow(screen)
```
